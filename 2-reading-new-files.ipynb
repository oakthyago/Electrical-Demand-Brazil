```python
def csv_compact_file (totalfiles,given_date,file_name,string):
    n=1
    date_format = '%m/%d/%Y'
    df2=pd.DataFrame()
    dtObj = datetime.strptime(given_date, date_format)
    for n in range(totalfiles):
        future_date = dtObj + relativedelta(days=n)
        x=future_date.strftime("%m")
        y=future_date.strftime("%d")
        #string = '../input/solarenergynortheastbrazil/2 Lindalva e Severino/2022-'+ x +"-"+ y +'_2022-' + x +"-"+ y +'InverterSK1ES175N54430Report.xlsx'
        df=pd.read_excel(string)
        df=df[['Time','Production（kW）','Consumption（kW）','Temperature（℃）','Weather']]
        df2=pd.concat([df2,df], ignore_index=True)
        
        df2['Date'] = df2['Time'].apply(lambda x: x.split(' ')[0])
        df2['Hour'] = df2['Time'].apply(lambda x: x.split(' ')[1])
        df2['Hour'] = pd.to_datetime(df2['Hour'], format='%H:%M')
        df2['Date'] = pd.to_datetime(df2['Date'], format='%Y/%m/%d')
        
    return(df2)
```
