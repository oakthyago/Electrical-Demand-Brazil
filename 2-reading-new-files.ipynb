```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
def csv_compact_file (totalfiles,given_date,file_name,a,c):
    n=1
    date_format = '%m/%d/%Y'
    df2=pd.DataFrame()
    dtObj = datetime.strptime(given_date, date_format)
    for n in range(totalfiles):
        future_date = dtObj + relativedelta(days=n)
        x=future_date.strftime("%m")
        y=future_date.strftime("%d")
        string = a + x +"-"+ y +'_2022-' + x +"-"+ y +c
        try:
            df=pd.read_excel(string)
        except:
            pass #df=df[['Time','Production（kW）','Consumption（kW）','Temperature（℃）','Weather']]
        df2=pd.concat([df2,df], ignore_index=True)
        
        #df2['Date'] = df2['Time'].apply(lambda x: x.split(' ')[0])
        #df2['Hour'] = df2['Time'].apply(lambda x: x.split(' ')[1])
        #df2['Hour'] = pd.to_datetime(df2['Hour'], format='%H:%M')
        #df2['Date'] = pd.to_datetime(df2['Date'], format='%Y/%m/%d')
        
    return(df2)
```

```python

Paulino=pd.DataFrame()

given_date = '05/06/2022'
totalfiles = 180
file_name = "Paulino"
a= '../input/solarenergynortheastbrazil/2 Paulino/2022-'
c='InverterSH3ES250MB8301Report.xlsx'
#--

Sevem =pd.DataFrame()
given_date2 = '08/12/2022'
totalfiles2 = 82
file_name2 = "Sevem e Marcelo"
a2= '../input/solarenergynortheastbrazil/2 Sevem/2022-'
c2= 'InverterSH3ES250MCV785Report.xlsx'

#---

Tubarao=pd.DataFrame()
given_date3 = '09/22/2022'
totalfiles3 = 40
file_name3 = "Tubarao"
a3= '../input/solarenergynortheastbrazil/2 Tubarao/2022-'
c3='InverterSK1ES175N54731Report.xlsx'
#----

Vinicius1=pd.DataFrame()
given_date4 = '05/22/2022'
totalfiles4 = 175
file_name4 = "Vinicius1"
a4 = '../input/solarenergynortheastbrazil/21 Vinicius/2022-'
c4 = 'InverterSH1ES160M39850Report.xlsx'

Vinicius2=pd.DataFrame()
given_date6 = '05/22/2022'
totalfiles6 = 178
file_name6 = "Vinicius2"
a6 = '../input/solarenergynortheastbrazil/22 Vinicius/2022-'
c6 = 'InverterSH1ES160M39783Report.xlsx'


Lindalva=pd.DataFrame()

a5 = '../input/solarenergynortheastbrazil/2 Lindalva e Severino/2022-'
c5='InverterSK1ES175N54430Report.xlsx'
given_date5 = '09/30/2022'
totalfiles5 = 30
file_name5 = "Lindalva e Severino"
```

```python
Paulino = csv_compact_file(totalfiles,given_date,file_name,a,c)
Sevem= csv_compact_file(totalfiles2,given_date2,file_name2,a2,c2)
Tubarao = csv_compact_file(totalfiles3,given_date3,file_name3,a3,c3)
Vinicius1= csv_compact_file(totalfiles4,given_date4,file_name4,a4,c4)
Lindalva= csv_compact_file(totalfiles5,given_date5,file_name5,a5,c5)
Vinicius2= csv_compact_file(totalfiles6,given_date6,file_name6,a6,c6)
tudo =pd.concat([Paulino,Sevem,Tubarao,Vinicius1,Vinicius2,Lindalva], ignore_index=True)
Vinicius_tudo =pd.concat([Vinicius1,Vinicius2], ignore_index=True)
```

```python
Paulino.to_csv("2_Paulino_newdata.csv")
Sevem.to_csv("2_Sevem_Marcelo_new_data.csv")
Tubarao.to_csv("2_Tubarao_new_data.csv")
Vinicius1.to_csv("Vinicius1.csv")
Vinicius2.to_csv("Vinicius2.csv")
Lindalva.to_csv("2_Lindalva_new_data.csv")
tudo.to_csv("All_Inverters.csv")
Vinicius_tudo.to_csv("All_Vinicius.csv")
```
